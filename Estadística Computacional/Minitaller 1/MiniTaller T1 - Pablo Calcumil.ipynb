````python
# Librerias necesarias para trabajar

import numpy as np
import matplotlib.pyplot as plt

import wget
import csv

# URL's de los datos

url_c1 = "https://raw.githubusercontent.com/FMena14/ML_usm/master/Estadistica/T1-Eval-Data/Muestra_C1.txt"
url_c2 = "https://raw.githubusercontent.com/FMena14/ML_usm/master/Estadistica/T1-Eval-Data/Muestra_C2.txt"
url_c3 = "https://raw.githubusercontent.com/FMena14/ML_usm/master/Estadistica/T1-Eval-Data/Muestra_C3.txt"

# Archivos de los datos

file_c1 = wget.download(url_c1)
file_c2 = wget.download(url_c2)
file_c3 = wget.download(url_c3)

c_1, c_2, c_3 = [], [], []

# Obtención de datos

with open('Muestra_C1.txt') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        c_1.append(float(row[0]))
        
with open('Muestra_C2.txt') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        c_2.append(float(row[0]))
        
with open('Muestra_C3.txt') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        c_3.append(float(row[0]))
````

> g) [15 pts] Ahora se le solicita realizar el estudio de qué ocurre si la empresa decide gradualmente enviar a sus mejores trabajadores (los con desempeño más alto) al trabajo presencial de oficina.  Compare cómo se ve afectada la media muestral de cada condición si un porcentaje $p$ de los trabajadores bajo esa condición son enviados a trabajo presencial (ya no formarían parte de esta muestra). Grafique la variación de $p$ en un rango $[0;0.99]$. En base a esto ¿Cuál condición sería recomendable mantener en trabajo online para evitar una baja en las medidas de tendencia central de trabajo online?


````python
def online(porcentaje, condicion, C1, C2, C3):
    if condicion == 1:
        lista = C1
    elif condicion == 2:
        lista = C2
    else:
        lista = C3
    lista.sort()
    lista = lista[0: int(len(lista) * (1 - porcentaje))]
    return lista
````

````python
import random as rd

#Obtencion de las medias

p = np.arange(0.00, 0.99, 0.01)
medias, medianas = [], []
for i in p:
    medias.append(np.mean(np.array(online(i, 1, c_1, c_2, c_3))))
    medianas.append(np.median(np.array(online(i, 1, c_1, c_2, c_3))))

#Grafico

fig, ax = plt.subplots(figsize = (12, 6))
ax.plot(p, medias)
ax.plot(p, medianas)
ax.set(xlabel = 'Porcentaje de Trabajadores Presencial',
       ylabel = 'Medidas de Tendencia Central',
       title = 'Medidas de Tendencia Central vs Porcentaje Trabajadores Presencial C1')
ax.grid()
plt.show()
````
