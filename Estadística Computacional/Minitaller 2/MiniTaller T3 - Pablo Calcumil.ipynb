## Introducción

El *cachipún* es un juego de manos muy practicado en nuestro país en el que existen tres movimientos posibles: la *piedra* que vence a la *tijera* rompiéndola, la *tijera* que vence al *papel* cortándolo, y el *papel* que vence a la *piedra* envolviéndola. En una ronda estándar de cachipún, dos jugadores cuentan juntos "1, 2, 3 ..." o directamente ¡ca, chi, pun! y justo al terminar muestran los dos al mismo tiempo una de sus manos, de modo que pueda verse el movimiento que cada uno ha elegido jugar y pueda determinarse el ganador.

1.   Determine experimentalmente la función de probabilidad $f(x)$ y el valor esperado $\mathbb{E}(X)$ de la recompensa $X$ que obtendrá en un turno. Compare los valores experimentales con los valores teóricos y comente. ¿Es correcto afirmar que $\mathbb{E}(X)$ es el valor más probable de la recompensa que obtendrá? ¿Es correcto afirmar que si juega con $10$ adversarios que ejecutan la misma estrategia su recompensa esperada es $10 \cdot \mathbb{E}(X)$?


**Respuesta**: Notemos que la variable aleatoria es $X$, que se define como la recompensa que se obtendrá en un turno, entonces los valores que esta variable toma son:
$$X = \begin{cases} 5000\hspace{0.7cm},\text{ gana piedra - tijera}\\ 3000\hspace{0.7cm},\text{ gana tijera - papel}\\ 1000\hspace{0.7cm},\text{ gana papel - piedra}\\ 0   \hspace{1.38cm},\text{ empate piedra, papel o tijera}\\ -5000\hspace{0.4cm},\text{ pierde tijera - piedra}\\ -3000\hspace{0.4cm},\text{ pierde papel - tijera}\\ -1000\hspace{0.4cm},\text{ pierde piedra - papel} \end{cases}$$


```python
def cachipun(n):                                # n es el numero de veces a realizar el cachipun
    games = []                                  # 0: Piedra, 1: Tijera, 2: Papel
    for i in range(0, n):                       # Son para ver el historial completo de las jugadas
        games.append((np.random.randint(3),np.random.randint(3))) # Posicion 0 soy yo
    suma_j1 = 0
    contadores = [0,0,0,0,0,0,0]                # Son para formar la probabilidad de cada caso
    for i in range(0,n):
        
        #Casos Ganados
        if games[i][0] == 0 and games[i][1] == 1:
            suma_j1 = suma_j1 + 5000
            contadores[0] = contadores[0] + 1
        elif games[i][0] == 1 and games[i][1] == 2:
            suma_j1 = suma_j1 + 3000
            contadores[1] = contadores[1] + 1
        elif games[i][0] == 2 and games[i][1] == 0:
            suma_j1 = suma_j1 + 1000
            contadores[2] = contadores[2] + 1
            
        #Casos Perdidos
        elif games[i][0] == 1 and games[i][1] == 0:
            suma_j1 = suma_j1 - 5000
            contadores[4] = contadores[4] + 1
        elif games[i][0] == 2 and games[i][1] == 1:
            suma_j1 = suma_j1 - 3000
            contadores[5] = contadores[5] + 1
        elif games[i][0] == 0 and games[i][1] == 2:
            suma_j1 = suma_j1 - 1000
            contadores[6] = contadores[6] + 1
        
        #Empate
        else:
            contadores[3] = contadores[3] + 1
            
    return suma_j1, contadores
```

```python
# Se saca los valores de la funcion y la esperanza experimentalmente

n = 10000

expected_v, contadores = cachipun(n)

print('La funcion de probabilidad es igual a:\n')
print('       ', round(contadores[0] / n, 4), 'si X = 5000')
print('       ', round(contadores[1] / n, 4), 'si X = 3000')
print('       ', round(contadores[2] / n, 4), 'si X = 1000')
print('f(X)=  ', round(contadores[3] / n, 4), 'si X = 0')
print('       ', round(contadores[4] / n, 4), 'si X = -5000')
print('       ', round(contadores[5] / n, 4), 'si X = -3000')
print('       ', round(contadores[6] / n, 4), 'si X = -1000')
print('\nLa esperanza de X es igual a:\n')
print('E[X] =', round(expected_v / n, 4))
```

Notamos que los valores de la variable aleatoria $X$ son 7, por ello de alguna manera, al considerar que lo que jugará cada jugador es independiente de los juegos anteriores, se puede decir que son equiprobables, pero se tiene que recordar que el valor $0$ tiene $3$ casos, que son empatar con piedra o con tijera, o con papel. Dicho todo esto, y al recordar que son equiprobables lo que jugarán se tiene que para cada caso teóricamente hay $\frac{1}{9}$ de probabilidades, excepto para el empate, ya que son $3$ casos, entonces hay $\frac{1}{3}$ de probabilidad de empatar. Por lo tanto, al ver que los valores encontrados experimentalmente son muy parecidos, se está abordando de manera correcta y con la debida comprensión el problema y la distribución de la variable aleatoria $X$.

Dado que se tiene que la probabilidad de que se empate es de $\frac{1}{3}$, y vemos que el valor de la esperanza es cercano a este valor de recompensa $0$, lo que es obvio dado que los valores de $X$ son equiprobables (a excepción del $X=0$), se puede decir que $E[X]$ es el valor de recompensa más probable.

Notemos que si jugamos contra $10$ personas, las cuales realizan la misma jugada, la variable aleatoria sería:

$$Y = \begin{cases} 50000\hspace{0.7cm},\text{ gana piedra - tijera}\\ 30000\hspace{0.7cm},\text{ gana tijera - papel}\\ 10000\hspace{0.7cm},\text{ gana papel - piedra}\\ 0   \hspace{1.56cm},\text{ empate piedra, papel o tijera}\\ -50000\hspace{0.4cm},\text{ pierde tijera - piedra}\\ -30000\hspace{0.4cm},\text{ pierde papel - tijera}\\ -10000\hspace{0.4cm},\text{ pierde piedra - papel} \end{cases}$$

Suponiendo que se debe de recibir y entregar la misma cantidad al ganar y al perder con cada juego. Notamos que $Y = 10 \cdot X$, y por propiedad de la esperanza, $E[Y] = 10 \cdot E[X]$.

Determine experimentalmente la función de probabilidad $f(z)$, el valor esperado $\mathbb{E}(Z)$, y la varianza $\mathbb{V}(Z)$ de la recompensa total $Z$ que obtendrá después de jugar 10 rondas. Presente la f.d.p. usando un gráfico de barras y comente. ¿Es correcto afirmar que $\mathbb{E}(Z)$ es el valor más probable de la recompensa que obtendrá? ¿Es cierto que $Z = 10 \cdot X$? ¿Es cierto que $\mathbb{E}(Z) = 10 \cdot \mathbb{E}(X)$? ¿Es cierto que $\mathbb{V}(Z) = 10 \cdot \mathbb{V}(X)$?


**Respuesta**: Es facil notar que la nueva variable aleatoria $Z$ no cumple con $Z = 10\cdot X$, ya que en este caso las 10 rondas se ocuparan distintas jugadas, por lo que es posible llegar a valores como $Z = 1000$. Es decir, este caso la variable aleatoria $Z$ tiene muchos mas valores a $X$.

```python
def cachipun_varios_juegos(n,m):            # n es el numero de veces a realizar el experimento, m veces cachipun 
    dicc = dict()
    suma_total, suma_total_cuadrada = 0, 0
    for i in range(0, n):
        suma, lista = cachipun(m)                        # Realizamos el cachipun 10 veces
        if str(suma) in dicc.keys():
            dicc[str(suma)] = dicc[str(suma)] + 1        # Contador de los valores al final de las 10 rondas
        else:
            dicc[str(suma)] = 1
        suma_total = suma_total + suma                            # La suma total de ganancias
        suma_total_cuadrada = suma_total_cuadrada + suma ** 2     # La suma total cuadrada de ganancias
    
    return suma_total, suma_total_cuadrada, dicc
```

```python
#Obteniendo probabilidades
expected, expected_2, diccionario  = cachipun_varios_juegos(n, 10)
values = list(diccionario.keys())
valores = [int(i) for i in values]
valores.sort()
values = [str(i) for i in valores]
probabilidades = [diccionario[values[i]] for i in range(0,len(values))]
probabilidades = np.array(probabilidades) / (n)
eje_y = np.array([i for i in range(0,51,2)]) / 1000

#Grafico

fig, ax = plt.subplots(figsize = (20, 10))
ax.bar(values, probabilidades)
plt.xticks(rotation = 55)
plt.yticks(eje_y)
plt.grid(linestyle = '--', axis = 'y', color = 'black')
plt.show()

print('\nLa esperanza de Z es igual a:\n')
print('E[Z] =', round(expected / n, 4))
print('\nLa varianza de Z es igual a:\n')
print('Var[Z] =', round(expected_2 / (n) - (expected / (n)) ** 2, 4))
print('\nLa esperanza de X es igual a:\n')
print('E[X] =', round(cachipun_varios_juegos(n,1)[0] / n, 4))
print('\nLa varianza de X es igual a:\n')
print('Var[X] =', round(cachipun_varios_juegos(n,1)[1] / (n) - (cachipun_varios_juegos(n,1)[0] / (n)) ** 2, 4))
```

Notamos que la variable aleatoria $Z$ puede tomar muchos valores, que pueden rondar desde los $-50000$ (perder todas), hasta los $50000$ (ganar todas), es decir, en total $Z$ puede tomar $101$ valores. Y con el gráfico se puede ver como está determinada la función de probabilidad $f(z)$. (La función facilmente se puede printear y escribir como en el caso anterior, ya que se tienen los valores que toma Z y las probabilidades [representadas en el grafico], pero esto abarcaría mucho espacio) y es mejor ver desde el gráfico).

Notamos también que las mayores probabilidades estan concentradas alrededor del $0$, lo que es lógico, ya que estos tienen mayor cantidad de formas para llegar a ellas, mientras que en las esquinas, como el valor $50000$, solo tiene una forma. Y con el $0$ que debe ser el que mayor probabilidad tiene de salir. Sumado a esto, tambien se puede notar que el grafico dado es muy simetrico, lo que concuerda con lo dicho, puesto que para un valor $X$ se tiene $r$ formas de formarlo, tambien hay $r$ formas de formar $-X$.



Por otro lado, gracias al mismo gráfico es claro qué $Z \not = 10 \cdot X$, y por propiedades de la esperanza, está claro que $E[Z] \not = 10 \cdot E[X]$, pero en este caso debiesen cumplirse, ya que como el gráfico debiese ser completamente simetrico, las esperanzas deben ser 0, pero notamos que experimentalmente no ocurrio así. 3.   Defina $T$ como el número de rondas en que usted y su oponente empatan después de jugar 10 veces. Determine la f.d.p, el valor esperado y la varianza de $T$ usando sólo papel, lápiz y pensamiento. Luego, use su simulación para verificar los resultados. Concluya.

**Respuesta**: Podemos notar que la variable que en cada juego tiene la probabilidad de $\frac{1}{3}$ de empatar, o de $\frac{2}{3}$ de ganar o perder, por lo que podemos considerar a $T$ como una variable aleatorial tal que $T \sim Bin(10,\frac{1}{3})$. Con esto se tiene que $E[T] = \frac{10}{3}$ y $Var[T] = \frac{20}{9}$. Y la funcion de probabilidad es:

\begin{align*}
f(t) &= 
\left\{
\begin{array}{cc}
{n\choose t}p^t(1-p)^{n-t}   &  \mbox{si } t = 1,2,\cdots,10   \\
0 &   \text{otro caso}  \\
\end{array}
\right.
 \ .
\end{align*}

```python
def empates(n, m):                                   # n para repetir experimento, m cantidad de rondas
    counts = np.zeros(7)
    results = np.zeros(11)
    ID = np.identity(11)
    suma_counts, suma_cuadrada = 0, 0
    for i in range(0, n):                               
        suma, lista = cachipun(m)                         # Rondas de cachipun 
        if lista[3] == 0:
            results = results + ID[0]                     # Sumamos en la posicion que corresponde
        elif lista[3] == 1:                               # la cantidad de empates que hubo
            results = results + ID[1]                     # posicion = cantidad de empates (incluyendo el 0)
        elif lista[3] == 2:
            results = results + ID[2]
        elif lista[3] == 3:
            results = results + ID[3]
        elif lista[3] == 4:
            results = results + ID[4]
        elif lista[3] == 5:
            results = results + ID[5]
        elif lista[3] == 6:
            results = results + ID[6]
        elif lista[3] == 7:
            results = results + ID[7]
        elif lista[3] == 8:
            results = results + ID[8]
        elif lista[3] == 9:
            results = results + ID[9]
        else:
            results = results + ID[10]
        counts = counts + np.array(lista)
        suma_counts = suma_counts + lista[3]                           #Esperanza y Varianza
        suma_cuadrada = suma_cuadrada + lista[3] ** 2
    return suma_counts, suma_cuadrada, results
```

```python
#Probabilidades

N = 100000
esperanza, esperanza_2, arreglo = empates(N,10)
arreglo = arreglo / N

print('La funcion de probabilidad es igual a:\n')
print('       ', round(arreglo[10], 4), '     si T = 10')
print('       ', round(arreglo[9], 4), '  si T = 9')
print('       ', round(arreglo[8], 4), '  si T = 8')
print('       ', round(arreglo[7], 4), '  si T = 7')
print('       ', round(arreglo[6], 4), '  si T = 6')
print('f(T)=  ', round(arreglo[5], 4), '  si T = 5')
print('       ', round(arreglo[4], 4), '  si T = 4')
print('       ', round(arreglo[3], 4), '  si T = 3')
print('       ', round(arreglo[2], 4), '  si T = 2')
print('       ', round(arreglo[1], 4), '  si T = 1')
print('       ', round(arreglo[0], 4), '  si T = 0')

print('\nLa esperanza de T es igual a:\n')
print('E[T] =', round(esperanza / N, 4))
print('\nLa varianza de T es igual a:\n')
print('Var[T] =', round(esperanza_2 / (N) - (esperanza / (N)) ** 2, 4))
```

```python
#Grafico

fig, ax = plt.subplots(figsize = (20, 10))
ax.bar(range(0, 11), arreglo)
plt.grid(linestyle = '--', axis = 'y', color = 'black')
plt.show()
```

Luego, viendo los valores que se presentaron, tanto experimentalmente como teóricamente, se puede concluir que exactamente la variable aleatoria $T$, sigue una distribución binomial. $T \sim Bin(10,\frac{1}{3})$.

4.   Suponga que usted decide continuar en el juego siempre y cuando no supere las 5 derrotas. Si definimos $S$ como el número de rondas jugadas, determine la f.d.p, el valor esperado y la varianza de $S$ usando sólo papel, lápiz y pensamiento. Luego, use su simulación para verificar los resultados. Concluya.

**Respuesta**: Podemos notar que la variable que cada juego tiene probabilidad $\frac{1}{3}$ de perder, y $\frac{2}{3}$ de ganar o empatar, por lo que se puede decir que la variable aleatoria $S$, esta distribuida de forma binomial negativa, es decir, $S \sim Bin^{-1}(5,\frac{1}{3})$. Con esto dicho, se tiene que $E[S] = 10$, $V[S] = 30$, y la función de probabilidad queda:

\begin{align*}
f(s) &= 
\left\{
\begin{array}{cc}
{s-1\choose 4}(1-p)^{s-5}p^5   &  \mbox{si } s = 5, 6, \cdots
\\
0 &  \text{otro caso}  \\
\end{array}
\right.
 \ .
\end{align*}

```python
def max_n(n,N):                                       # N veces se repite experimento, n maximo de fallas
    rondas = dict()
    suma, suma_cuadrada = 0, 0
    for i in range(0, N):                                            # Repeticion del experimento
        count_loss = 0
        count_rondas = 0
        while count_loss != 5:                                        # Se detiene al llegar a las 5 perdidas
            _, lista = cachipun(1)
            if lista[4] == 1 or lista[5] == 1 or lista[6] == 1:      # Posiciones de las perdidas en cachipun()
                count_loss = count_loss + 1
            count_rondas = count_rondas + 1
        if str(count_rondas) not in rondas.keys():                    # rondas es dict que cuenta las rondas  
            rondas[str(count_rondas)] = 1                             # que se realizan hasta llegar a la 5ta
        else:                                                         # vez que se pierde
            rondas[str(count_rondas)] = rondas[str(count_rondas)] + 1  # el numero de rondas es la llave y
        suma = suma + count_rondas                                     # el valor es las veces que sucede lo mismo
        suma_cuadrada = suma_cuadrada + count_rondas ** 2              #Esperanza y Varianza
    
    return suma, suma_cuadrada, rondas
```

```python
#Probabilidades

esperanza, esperanza_2, dicc = max_n(5,N)
values = list(dicc.keys())
valores = [int(i) for i in values]
valores.sort()
probabilidades = [dicc[str(valores[i])] for i in range(0,len(valores))]
values = [str(i) for i in valores]
probabilidades = np.array(probabilidades) / (N)
eje_y = np.array([i for i in range(0,101,2)]) / 1000

#Grafico

fig, ax = plt.subplots(figsize = (20, 10))
ax.bar(values, probabilidades)
plt.xticks(rotation = 55)
plt.yticks(eje_y)
plt.grid(linestyle = '--', axis = 'y', color = 'black')
plt.show()

print('La funcion de probabilidad de S es igual a:\n')
for i in range(0, len(valores)):
    print(probabilidades[i], 'si S = ', valores[i])

print('\nLa esperanza de S es igual a:\n')
print('E[S] =', round(esperanza / N, 4))
print('\nLa varianza de S es igual a:\n')
print('Var[S] =', round(esperanza_2 / (N) - (esperanza / (N)) ** 2, 4))
```

Podemos notar que los valores dieron muy parecidos, tanto para la función de probabilidad, como para la esperanza y la varianza. Con esto es correcto decir que la variable aleatoria $S$ es tal que $S \sim Bin^{-1}(5,\frac{1}{3})$.

5.   Suponga que usted y su oponente pactan iterar el juego hasta que la recompensa total de cualquiera de los jugadores exceda los 10.000 pesos. Si definimos $R$ como el número de rondas jugadas, use su simulación para determinar experimentalmente la f.d.p, el valor esperado y la varianza de $R$.

**Respuesta**: Se forma la función `rondas_10k()` para ver la función de probabilidad, la esperanza y la varianza de $R$.

```python
def rondas_10k(N):                                          # N veces que se repite el experimento
    rondas = dict()
    suma_esp, suma_var = 0, 0
    for i in range(0, N):                                   # Repetir el experimento
        count_10k = 0
        count_rondas = 0
        while count_10k < 10000 and count_10k > -10000:      # Si llego a 10k termina, si llego a -10k tambien,
            suma, _ = cachipun(1)                            # pues eso quiere decir que el lleva 10k
            count_rondas = count_rondas + 1
            count_10k = count_10k + suma
        if str(count_rondas) not in rondas.keys():                    # rondas es dict que cuenta las rondas  
            rondas[str(count_rondas)] = 1                             # que se realizan hasta llegar a los |10000|
        else:                                                         # vez que se pierde
            rondas[str(count_rondas)] = rondas[str(count_rondas)] + 1  # el numero de rondas es la llave y
        suma_esp = suma_esp + count_rondas                           # el valor es las veces que sucede lo mismo
        suma_var = suma_var + count_rondas ** 2
    
    return suma_esp, suma_var, rondas
```

```python
# Probabilidades

esperanza, esperanza_2, dicc = rondas_10k(N)
values = list(dicc.keys())
valores = [int(i) for i in values]
valores.sort()
probabilidades = [dicc[str(valores[i])] for i in range(0,len(valores))]
values = [str(i) for i in valores]
probabilidades = np.array(probabilidades) / (N)
eje_y = np.array([i for i in range(0,59,2)]) / 1000

#Grafico

fig, ax = plt.subplots(figsize = (20, 10))
ax.bar(values, probabilidades)
plt.xticks(rotation = 55)
plt.yticks(eje_y)
plt.grid(linestyle = '--', axis = 'y', color = 'black')
plt.show()

print('La funcion de probabilidad de S es igual a:\n')
for i in range(0, len(valores)):
    print(probabilidades[i], 'si S = ', valores[i])

print('\nVerificamos que las probabilidades suman:', np.sum(probabilidades))

print('\nLa esperanza de R es igual a:\n')
print('E[R] =', round(esperanza / N, 4))
print('\nLa varianza de R es igual a:\n')
print('Var[R] =', round(esperanza_2 / (N) - (esperanza / (N)) ** 2, 4))
```

Se puede ver que hay casos en que se llegan a jugar mas de 100 rondas para poder llegar a los $10000$ ganados o perdidos, pero la probabilidad de llegar ahí es más baja que la que se tiene al jugar 2 partidas y ganar (o perder) los $10000$.

**Respuesta**:

Tomaremos la estrategia donde se utiliza piedra con probabilidad $\frac{7}{10}$, papel con $\frac{3}{10}$ y nunca se ultiliza tijeras, esto podría notar que se puede perder en algunos casos, pero este resultaria en menores perdidas, por lo que no se puede asegurar siempre recompensas positivas en los 10 juegos. 

Por otro lado, teóricamente se tiene que el valor esperado de un solo juego resultaría:

$$E[X] = \frac{7}{10}\cdot \frac{1}{3}\cdot 5000 +\frac{7}{10}\cdot \frac{1}{3}\cdot (-1000) +\frac{7}{10}\cdot \frac{1}{3}\cdot 0 +\frac{3}{10}\cdot \frac{1}{3}\cdot (-3000) +\frac{3}{10}\cdot \frac{1}{3}\cdot 1000 +\frac{3}{10}\cdot \frac{1}{3}\cdot 0 = 733.3$$

Pero sabemos, por lo dicho anteriormente, que en este caso la variable aleatoria de los 10 juegos, llamemosla $Z$, cumple $Z\not = 10 \cdot X$, y por ello no se debería cumplir que $E[Z] = 10 \cdot E[X]$.

```python
def cachipun_estrategia(n):                                # n es el numero de veces a realizar el cachipun
    games = []                                  # 0: Piedra, 1: Papel, 2: Tijera
    for i in range(0, n):                       # Son para ver el historial completo de las jugadas
        games.append((bernoulli.rvs(0.3),np.random.randint(3))) # Bernoulli(0,3), es decir es 1 con prob 0.3
    suma_j1 = 0
    contadores = [0,0,0,0,0,0,0]                # Son para formar la probabilidad de cada caso
    for i in range(0,n):
        
        #Casos Ganados
        if games[i][0] == 0 and games[i][1] == 2:
            suma_j1 = suma_j1 + 5000
            contadores[0] = contadores[0] + 1
        elif games[i][0] == 1 and games[i][1] == 0:
            suma_j1 = suma_j1 + 1000
            contadores[2] = contadores[2] + 1
            
        #Casos Perdidos
        elif games[i][0] == 1 and games[i][1] == 2:
            suma_j1 = suma_j1 - 3000
            contadores[5] = contadores[5] + 1
        elif games[i][0] == 0 and games[i][1] == 1:
            suma_j1 = suma_j1 - 1000
            contadores[6] = contadores[6] + 1
        
        #Empate
        else:
            contadores[3] = contadores[3] + 1
            
    return suma_j1, contadores

def cachipun_varios_juegos_e(n,m):            # n es el numero de veces a realizar el experimento, m veces cachipun 
    dicc = dict()
    suma_total, suma_total_cuadrada = 0, 0
    for i in range(0, n):
        suma, lista = cachipun_estrategia(m)                        # Realizamos el cachipun 10 veces
        if str(suma) in dicc.keys():
            dicc[str(suma)] = dicc[str(suma)] + 1        # Contador de los valores al final de las 10 rondas
        else:
            dicc[str(suma)] = 1
        suma_total = suma_total + suma                            # La suma total de ganancias
        suma_total_cuadrada = suma_total_cuadrada + suma ** 2     # La suma total cuadrada de ganancias
    
    return suma_total, suma_total_cuadrada, dicc
```

```python
#Obteniendo probabilidades
expected, expected_2, diccionario  = cachipun_varios_juegos_e(n, 10)
vals = list(diccionario.keys())
valos = [int(i) for i in vals]
valos.sort()
vals = [str(i) for i in valos]
probs = [diccionario[vals[i]] for i in range(0,len(vals))]
probs = np.array(probs) / (n)
eje_y = np.array([i for i in range(0,70,5)]) / 1000

#Grafico

fig, ax = plt.subplots(figsize = (20, 10))
ax.bar(vals, probs)
plt.xticks(rotation = 55)
plt.yticks(eje_y)
plt.grid(linestyle = '--', axis = 'y', color = 'black')
plt.show()

print('\nLa esperanza de Z es igual a:\n')
print('E[Z] =', round(expected / n, 4))
```

A través del gráfico podemos notar que, si bien no se logran solo recompensas positivas tras los 10 juegos, la probabilidad de obtener una recompensa mayor a $0$, es mucho mayor a la de tener recompensa menor o igual a 0.

Por otro lado, podriamos decir (notando el valor dado como esperanza de $Z$) que de cierta manera se cumple $E[Z] = 10 \cdot E[X]$, y que el valor esperado de recompensa tras los 10 juegos, es positivo.
